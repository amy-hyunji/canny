```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as Ftnl

%matplotlib inline
import matplotlib.pyplot as plt
def graph(x,f,plot=True):
    if not plot:
        return (x,f(x))
    plt.plot(x,f(x))
    return (x,f(x))
```

```python
def f(x,degree=10):
    return torch.clamp(x,0,1) ** degree

def g(x,degree=10):
    return 1 - (torch.clamp(x,0,1) ** degree + torch.clamp(-x+1,0,1) ** degree)

def h(x,degree=10):
    return torch.clamp(-x+1,0,1) ** degree

X = torch.linspace(-2,2,10000)

plt.figure()
x, fx = graph(X,f)
x, gx = graph(X,g)
x, hx = graph(X,h)
pf = fx/(fx+gx+hx)
pg = gx/(fx+gx+hx)
ph = hx/(fx+gx+hx)
plt.figure()
plt.plot(x,pf)
plt.plot(x,pg)
plt.plot(x,ph)
plt.plot(x,pf+pg+ph)
```
